# Features


```python
def compute_ptp_amp(data):
    return np.ptp(data, axis=-1)
```

```python
def compute_rms(data):
    return np.sqrt(np.mean(np.power(data, 2), axis=-1))
```

```python

F_ptp=[]
F_rms=[]
F_std=[]
F_max=[]


F_ptp_test=[]
F_rms_test=[]
F_std_test=[]
F_max_test=[]
for i in range(16):

  F_ptp.append( compute_ptp_amp(data_x_train[i]))
  F_rms.append( compute_rms(data_x_train[i]))
  F_std.append( np.std(data_x_train[i], axis=2))
  F_max.append(  np.max(data_x_train[i], axis=2))


for i in range(5):

  F_ptp_test.append( compute_ptp_amp(data_x_test[i]))
  F_rms_test.append( compute_rms(data_x_test[i]))
  F_std_test.append( np.std(data_x_test[i], axis=2))
  F_max_test.append(  np.max(data_x_test[i], axis=2))
```

```python
len(F_ptp)
```

```python
F_ptp_conc=np.concatenate(F_ptp)
F_rms_conc=np.concatenate(F_rms)
F_std_conc=np.concatenate(F_std)
F_max_conc=np.concatenate(F_max)

F_ptp_conc_test=np.concatenate(F_ptp_test)
F_rms_conc_test=np.concatenate(F_rms_test)
F_std_conc_test=np.concatenate(F_std_test)
F_max_conc_test=np.concatenate(F_max_test)
```
